# DATA Balacing

````python
data1 = pd.read_csv('hotel_booking.csv')
````

````python
x=data1.drop('is_canceled',axis =1)
y=data1['is_canceled']
from imblearn.under_sampling import RandomUnderSampler
rus=RandomUnderSampler(random_state=20,replacement=True)
````

````python
#x.head()
x_rus,y_rus = rus.fit_resample(x,y)
#print(x_rus,y_rus)
````

````python
data1.columns
````

````python
col =['lead_time', 'hotel',
'arrival_date_month', 'arrival_date_year',
'stays_in_weekend_nights',
'stays_in_week_nights',
'adults','children', 'babies', 'country',
'meal',
'market_segment',
'distribution_channel',
'is_repeated_guest',
'previous_cancellations',
'previous_bookings_not_canceled',
'booking_changes',
'deposit_type',
'customer_type',
'adr']
````

````python
#Taking only important features
df = x_rus[col]
df.shape
````

````python
df['is_canceled'] = y_rus
df.head()
````

# model training

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV

from sklearn.metrics import confusion_matrix, balanced_accuracy_score, classification_report
````

````python
df.info()
````

````python
df.dropna(axis = 0,inplace=True)
````

````python
x = df.drop('is_canceled',axis = 1)
y = df['is_canceled']
````

````python
y.value_counts()
````

````python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
````

````python
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
````

# Random Forest

````python
# Testing Data Evaluation
y_pred = rfc.predict(x_test)

cnf_matrix = confusion_matrix(y_test,y_pred)
print("Confusion Matrix", cnf_matrix)

acc_score = balanced_accuracy_score(y_test,y_pred)
print("Accuracy score", acc_score)

clf_report = classification_report(y_test,y_pred)
print("Classification report", clf_report)
````

````python
# Training Data Evaluation
y_pred = rfc.predict(x_train)

cnf_matrix = confusion_matrix(y_train,y_pred)
print("Confusion Matrix", cnf_matrix)

acc_score = balanced_accuracy_score(y_train,y_pred)
print("Accuracy score", acc_score)

clf_report = classification_report(y_train,y_pred)
print("Classification report", clf_report)
````

# Hyper parameter tuning RandomForest

````python
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
````

````python
rfc = RandomForestClassifier(random_state=30)

hyp = {"n_estimators" : np.arange(10,150),
    "criterion":['gini','entropy'],
    "max_depth":np.arange(5,15),
    "min_samples_split":np.arange(5,20),
    "min_samples_leaf":np.arange(5,15)}

rscv = RandomizedSearchCV(rfc, hyp, cv =5)
rscv.fit(x_train,y_train)
rscv.best_estimator_
````
